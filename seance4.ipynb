```python
label=df.iloc[:,0]
x=df.iloc[:,1:]
x.shape
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
sc.fit(x)
Xsc=sc.transform(x)
```

```python
pca=PCA()
x_pca = pca.fit_transform(Xsc)
# pca.components_
```

```python
pca.explained_variance_ratio_
```

```python

plt.figure(figsize=(10,10))
plt.scatter(x_pca[:,0],x_pca[:,1])
for i,txt in enumerate (label):
    plt.annotate(txt, (x_pca[:,0][i], x_pca[:,1][i]))
plt.xlabel('pca1')
plt.ylabel('pca2')
```

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator
```

```python
wcss = []
plt.figure(figsize=(10,10))

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kneedle = KneeLocator(range(1,11),wcss, curve='convex', direction='decreasing')
print("K={}".format(kneedle.elbow))

kneedle.plot_knee()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
kmeans.fit(x)
```

```python
plt.figure(figsize=(15,15))
plt.scatter(x_pca[:, 0], x_pca[:, 1],c=kmeans.labels_)
for i,txt in enumerate (label):
    plt.annotate(txt, (x_pca[:,0][i], x_pca[:,1][i]))
plt.xlabel('PCA1')
plt.ylabel('PCA2')
```

```python
kmeans_scaled = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
kmeans_scaled.fit(Xsc)
```

```python
plt.figure(figsize=(15,15))
plt.scatter(x_pca[:, 0], x_pca[:, 1],c=kmeans_scaled.labels_)
for i,txt in enumerate (label):
    plt.annotate(txt, (x_pca[:,0][i], x_pca[:,1][i]))
plt.xlabel('PCA1')
plt.ylabel('PCA2')
```

```python
kmeans_pca12 = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
kmeans_pca12.fit(x_pca[:,:2])
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(25, 10))
ax1.set_title("x_pca12")
ax1.scatter(x_pca[:, 0], x_pca[:, 1],c=kmeans_pca12.labels_)
for i in range (0,50):
    ax1.annotate(label[i][:3], (x_pca[i,0],x_pca[i,1]))
ax2.set_title("x_sc")
ax2.scatter(x_pca[:, 0], x_pca[:, 1],c=kmeans_scaled.labels_)
for i in range (0,50):
    ax2.annotate(label[i][:3], (x_pca[i,0],x_pca[i,1]))

fig.show()
```
